# Packages

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

# Dataset

```python
df = pd.read_csv("/content/carbon_emissions_by_distance.csv")
df
```

# Train

```python
X = df[['distance (km)']]
y_car = df['car (kg CO2)']

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_train_scaled, y_car, test_size=0.3, random_state=42)

model_carbon_car = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1)
])

model_carbon_car.compile(optimizer='adam',
              loss='mean_squared_error')

history = model_carbon_car.fit(X_train,
                    y_train,
                    epochs=100,
                    batch_size=32,
                    validation_data=(X_test, y_test))

loss = model_carbon_car.evaluate(X_test, y_test)
print(f"loss: {loss}")
```

```python
model_carbon_car.save("model_carbon_car.h5")
```
