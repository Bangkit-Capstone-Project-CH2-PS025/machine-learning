```python
import uuid
import pandas as pd
import numpy as np
```

```python
categories = ["Alam", "Pegunungan", "Sejarah dan Budaya", "Tempat Rekreasi",
              "Tempat Ibadah", "Desa Wisata", "Perbelanjaan", "Edukasi"]

np.random.seed(42)
user_data = []

for _ in range(1, 601):
    user_id = str(uuid.uuid4())  # Random UUID
    user_categories = np.random.choice(categories, size=2, replace=False)
    user_data.append([user_id, user_categories[0], user_categories[1]])

df = pd.DataFrame(user_data, columns=['user_id', 'preference_1', 'preference_2'])

df.set_index('user_id', inplace=True)

print(df)
```

```python
np.random.seed(42)
user_budget_data = []

for user_id in range(1, 401):
    # Random budget between 0 and 5,000,000 rounded to the nearest thousand
    budget = np.random.randint(0, 5001) * 1000
    user_budget_data.append([budget])

df2 = pd.DataFrame(user_budget_data, columns=['budget'])

print(df2)
```

```python
np.random.seed(42)
user_duration_data = []

for user_id in range(1, 401):
    duration = np.random.randint(1, 12)  # Random duration between 1 and 11
    user_duration_data.append([duration])

df3 = pd.DataFrame(user_duration_data, columns=['duration'])

print(df3)
```

```python
merged_df = df.merge(df2, left_index=True, right_index=True).merge(df3, left_index=True, right_index=True)

print(merged_df)
```
